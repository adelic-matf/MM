# Zakoni održanja

Krenimo prvo od nekih osnovnih fizičkih pojmova. **Fizički sistem** je skup objekata, sila i procesa koji su u međusobnoj interakciji i koji će predstavljati predmet izučavanja. Da bi se opisao fizički sistem mora se definisati šta je njegova granica, okolina, njegovi objekti i interakcije među njima.

Razlikujemo **izolovan**, **zatvoren** i **otvoren** sistem.

**Izolovan** sistem je onaj u kome nema razmene ni mase ni energije sa okolinom, dok
**zatvoreni sistem** dozvoljava razmene određenih veličina. Na primer, u termodinamici, u zatvorenom sistemu nema razmene mase sa okolinom, ali može postojati razmena energije.  U nekim problemima, termini izolovan i zatvoren sistem imaju isto značenje.

U **otvorenom** sistemu masa i energija mogu ulaziti u sistem i napuštati ga.

**Zakoni održanja** su osnovni principi fizike koji opisuju očuvanje određenih fizičkih veličina u zatvorenim ( ili izolovanim ) fizičkim sistemima. Zakoni kažu da se količina mase, energije ili impulsa, ne može stvoriti sama od sebe niti uništiti, već samo može preći iz jednog oblika u drugi.

**Primer - zakon održanja energije:** Najjednostavniji primer fizičkog sistema je kretanje klatna. Izvedimo klatno iz ravnotežnog položaja. Ako pretposavimo da se masa klatna ne menja i ako zanemarimo gubitak energije usled trenja, možemo primeniti zakon održanja energije (jer smo zatvorili sistem). Zakon kaže da se energija ne može sama od sebe stvoriti niti nestati unutar ovog sistema. U granicama ovog sistema ostavljamo silu gravitacije, jer bez nje ne bismo imali konverziju energije iz potencijalne u kinetičku. U momentu izvođenja klatna iz ravnotežnog položaja, klatno ima potencijalnu energiju, koja nakon puštanja klatna da se kreće, prelazi u kinetičku i obrnuto. Ukoliko ne dodajemo novu energiju sistemu (ne guramo klatno dodatno), ono ne može povećati amplitudu svog kretanja. Dok ako uključimo u sistem i otpor vazduha, ukupna energija u sistemu će se smanjivati sve dok ne istroši u potpunosti.

**Primer-zakon održanja mase:** Pretpostavimo da smo zapalili neko čvrsto gorivo, na primer drva. Zakon održanja mase kaže da materija ne može biti uništena niti stvorena u hemijskoj reakciji, već samo može preći u drugi oblik. Dakle, kada drvo gori, masa pepela, gasova i čađi koja se stvara jednaka je početnoj masi drva i kiseonika koji se nalazi u ovom fizičkom sistemu.

Zako održanja, u jednodimenzionom slučaju, se modeluju sledećom jednačinom
$$
\rho_t+q_x=0,
$$
gde je:
- $\rho$ gustina veličine koja se konzervira (ostaje održana)
- $q$ fluks, tj. protok fitizičke veličine kroz jediničnu površinu fizičkog sistema

Ova jednačina se može rešiti ako postoji neka nelinearna veza između fluksa i gustine, koju možemo zapisati sa
$$
q=F(\rho).
$$

Dobili smo **hiperboličku kvazilinearnu parcijalnu diferencijalnu jednačinu prvog reda** oblika
$$
\frac{\partial \rho}{\partial t}+\frac{\partial F(\rho)}{\partial x}=0,
$$

Ovaj tip jednačina se rešava **metodom karakteristika** koju ćemo ukratko objasniti.

## Metoda karakteristika

Neka je zadata funkcija $f:\mathbb{R}^n\to \mathbb{R} $, sa $y=f(x_1, x_2,\dots, x_n)$. Neka je $c$ konstanta. Označimo sa $\gamma$ nivoski skup tačaka funkcije $f(x)=c$.

```python
# grafik funkcije u 3D i nivockog skupa
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definišimo funkciju
def f(x, y):
    return x**2 + y**2

# Generišemo opseg vrednosti x i y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y) #formiramo mreze tacaka
Z = f(X, Y) #vrednosti funkcije na mrezi
c=4
# Prikazivanje 3D površine
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Prikazivanje površine
ax.plot_surface(X, Y, Z,alpha=0.6, cmap='viridis', edgecolor='none')

# Prikazivanje nivoskog skupova
contour = ax.contour3D(X, Y, Z, levels=[c], colors='black', linewidths=1)


# Dodavanje teksta u blizini kružnice
x_label=1
y_label=-1
z_label=4
ax.text(x_label, y_label, z_label, f"$\gamma: x^2 + y^2 = {c}$", color='red', fontsize=12)

# Podešavanje naslova i oznaka osa
ax.set_title(f"Grafik funkcje f(x,y)=x^2 + y^2 i njen nivoski skup $x^2 + y^2={c}$")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
#plt.colorbar(contour, label="Vrednosti nivoa")
ax.view_init(elev=20, azim=30)  # Upravite pogled 3D grafika

plt.show()
```

Parametrizujmo krivu $\gamma$ sa $x_i=x_i(s)$, $i=1,2,\dots, s$. Neka je $M$ tačka koja pripada krivoj $\gamma$ i neka je $r$ njen vektor položaja. Tada je pravac tangente na $\gamma$ dat sa
$$
\dot r=\left( \frac{dx_1}{ds},\frac{dx_2}{ds}, \dots, \frac{dx_n}{ds}\right)
$$

Kako je funkcija $f$ jednaka konstanti duž krive $\gamma$ to je njen izvod duz te krive jednak nuli:
$$
\frac{df}{ds}=0,
$$
tj.
$$
\frac{\partial f}{\partial x_1}\frac{\partial x_1}{\partial s}+\frac{\partial f}{\partial x_2}\frac{\partial x_2}{\partial s}+\cdots +
\frac{\partial f}{\partial x_n}\frac{\partial x_n}{\partial s}=0,
$$
što možemo zapisati pomoću skalarnog proizvoda
$$
\nabla f \cdot \dot r=0.
$$


Pogledajmo kako ovu činjenicu možemo iskoristi pri rešavanju parcijalnih diferencijalnih jednačina prvog reda. Neka je $u=u(x,t)$ rešenje parcijalne diferencijalne jednačine
$$
\frac{\partial u}{\partial t}+a\frac{\partial u}{\partial x}=0,
$$
što možemo zapisati i preko skalarnog proizvoda
$$
\left( \frac{\partial u}{\partial x},\frac{\partial u}{\partial t},-1\right)\cdot\left( a,1, 0\right) =0
$$
Definišimo funkciju $f(x,t,u)=u(x,t)-u$ i označimo sa $\gamma$ nivoski skup
$$
f(x,t,u)=u(x,t)-u=0.
$$
Očigledno on daje rešenje zadate pdj. Parametrizovanjem krive $\gamma$ i korišćenjem činjenice da je gradijent funkcije ortogonalan na tangentu krive nivoskog skupa dobijamo jednakost
$$
\left( \frac{\partial u}{\partial x},\frac{\partial u}{\partial t},-1\right)\cdot\left( \frac{dx}{ds},\frac{dt}{ds}, \frac{du}{ds}\right) =0
$$

Dalje izjednačavanjem ova dobijena dva skalarna proizvoda, umesto parcijalne diferencijalne jednačine, dobijamo sistem običnih diferencijalnih jednačina koji možemo lako rešiti:
$$
\frac{dx}{ds}=a,
$$
$$
\frac{dt}{ds}=1,
$$
$$
\frac{du}{ds}=0.
$$
Ovaj sistem jednačina nazivamo sistemom karakterističnih jednačina pdj. Njihova rešenja se nazivaju karakteristične krive ili karakteristike. Unijom karakterističnih krivih dobijamo rešenje polaznog problema.

## Jednačina linearne advekcije

Prvi i osnovni primer pdj prvog reda je jednačina linearne advekcije. Ona opisuje kretanje materije u nekom prostoru:
$$
\frac{\partial u}{\partial t}+a\frac{\partial u}{\partial x}=0.
$$
Ovde je:
- $u(x,t)$ veličina koja se transportuje (obično gustina materije)
- $a=const$ brzina kretanja  te veličine.
Rešenje izgleda poput talasa koji se kreće, na levo ako je $a<0$, odnosno na desno ako je $a>0$.

Rešimo ovu jednačinu metodom karakteristika. Već smo postavili sistem, pa jednačine treba samo da integralimo po promenljivoj $s$ čime dobijamo:
$$
x(s)=as+C_0
$$
$$
t(s)=s+\bar C_0,
$$
$$
u(s)=C_1.
$$

Eilimnacijom promenljive $s$ iz prve dve jednacine dobijamo karakteristike:
$$
x-at=C_2
$$
$$
u=C_1.
$$


Primetimo da kako je $a=const.$ to su projekcije karakterstika, tj. krive $x-at=C_2$ paralelne prave:

```python
## Projekcije karakteristika
# Definišemo parametre
a = 1  # brzina
C_values =np.arange(-5, 6)# razne vrednosti C2

# Generišemo vreme (t) i prostor (x)
t = np.linspace(0, 5, 100)  # vreme od 0 do 5
x_values = [a * t + C for C in C_values]  # x = at + C2

# Plotovanje grafika
plt.figure(figsize=(10, 6))
for x, C in zip(x_values, C_values):
    plt.plot( x,t, label=f'C2 = {C}')

# Dodavanje oznaka i legendi
plt.title('Grafik za x = at + C2 za razne vrednosti C2', fontsize=14)
plt.ylabel('t', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Osa x
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Osa y
plt.axis('equal')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

Duž ovih pravih rešenje je jednako konstati. Ali problem je što ne znamo kojoj. Postoji beskonačno mnogo načina na koji se mogu dovesti u vezu konstane $C_1=u$ i $C_2=x-at$, pa tu vezu jedino možemo izrazati nekom nepoznatom funkcijom $f$:
$$
C_1=f(C_2),
$$
tj. opšte rešenje zapisujemo kao
$$
u(x,t)=f(x-at).
$$

Međutim, ako je zadat početni uslov,
$$
u(x,0)=u_0(x)
$$
onda pdj ima jedinstveno rešenje, a kako iz opšteg rešenja sledi da je
$$
u(x,0)=f(x),
$$
to onda mora biti $f=u_0$, tj. jedinstveno rešenje je
$$
u(x,t)=u_0(x-at).
$$


**Primer** Neka je $a=1$ i neka je početni uslov zadat sa
$$
u_0(x)=e^{-x^2}.
$$
Rešenje jednačine linearne advekcije je tada $u(x,t)=e^{-(x-t)^2}$. Prikažimo grafik rešenja.

```python
# Definišemo opsege za x i t
x = np.linspace(-8, 8, 100)  # x od -5 do 5
t = np.linspace(0, 5, 100)   # t od 0 do 5

# Kreiramo mrežu tačaka (x, t)
X, T = np.meshgrid(x, t)

# Računamo vrednosti funkcije u(x, t) = exp((x - t)^2)
U = np.exp(-1*(X - T)**2)

# Kreiramo 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotovanje površine
ax.plot_surface(X, T, U, cmap='cividis', edgecolor='none')

# Dodavanje oznaka
ax.set_title('3D graf funkcije $u(x, t) = \exp(-(x - t)^2)$', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('t', fontsize=12)
ax.set_zlabel('u(x, t)', fontsize=12)

# Prikazivanje grafika
plt.show()
```

Rešenje možemo prikazati i koristeći karakteristične jednačine. Prikazaćemo sada rešenje kao animaciju, tj. za svako fiksirano t skiciraćemo grafik rešenja.

```python
def u0(x):
    return np.exp(-x**2)
xa=-5
xb=15
T=10;
C = np.linspace(xa, xb, 100)
t = np.linspace(0, T,101)
a=1 #brzina talasa

u_values=[]
x_values=[]

for i in range(len(t)):
    # izracunavamo vrednosti za x
    x = a * t[i] + C  # projekcije karakteristika za zadato vreme
    x_values.append(x)
    u = u0(C) #vrednosti funkcije na projekcijama karakteristika
    u_values.append(u)

## Animacija
#inicijalizacija figure
fig, axis = plt.subplots()
plt.axis([xa, xb, 0,1])
plt.xlabel('x')
plt.ylabel('u')
plt.show()


from matplotlib.animation import FuncAnimation
animplot, = axis.plot([], [])#grafički objekat koji će se menjati tokom animacije
#plot([], [], ...): Postavlja početne koordinate tačke na prazne vrednosti ([]) jer u početku nema podataka.


# Funkcija koja ažurira frejmove
def update_data(frame):
    animplot.set_data(x_values[frame],u_values[frame])  # Ažuriraj podatke
    axis.set_title(f"Brzina talasa a = {a}, t = {t[frame]:.2f}")  # Ažuriraj naslov
    return [animplot]

# Kreiranje animacije
anim = FuncAnimation(
    fig=fig,                   # pokazivac na figuru
    func=update_data,          # funkcija koja ažurira figuru, dodaje objekte na nju
    frames=range(0, len(x_values)),   # Broj frejmova
    interval=50,               # Pauza između frejmova (ms)
    blit=True,                 # # blit=True re-draws only the parts that have changed.
    repeat=False               # Ne ponavljaj animaciju
)

# Prikazivanje animacije
plt.show()
```
